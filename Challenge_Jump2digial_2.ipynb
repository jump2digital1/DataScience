```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_1 = df.drop(["created_at"],axis=1)
df_1 = df_1.drop(["send_at"],axis=1) 
df_1 = df_1.drop(["fechadevolucion"],axis=1) 
df_1 = df_1.drop(["money_back_date"],axis=1) 
df_1 = df_1.drop(["id"],axis=1) 
df_1 = df_1.drop(["user_id"],axis=1) 
df_1 = df_1.drop(["MONTH_ID"],axis=1) 

dummies = pd.get_dummies(df_1)
dummies
```

```python
dummies.fillna(dummies.mean(), inplace=True)
y = dummies['amount']
x = dummies.drop(['amount'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.sum().round(2)
```

```python
np.mean(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
sqrt(mean_squared_error(y_pred, y_test))
```
